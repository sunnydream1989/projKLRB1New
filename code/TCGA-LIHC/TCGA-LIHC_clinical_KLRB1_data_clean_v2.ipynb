# KLRB1

```python
if False:
    df_rnaseq = pd.read_csv("TCGA-LIHC\TCGA-LIHC-FPKM.symbol.txt", sep='\t')
    df_rnaseq.columns = map(lambda x: x.replace('.', '-'), df_rnaseq.columns)
    df_rnaseq.set_index('symbol',inplace=True)    
else:
    df_rnaseq = pd.read_csv("../../data/TCGA-LIHC/clean/TCGA-LIHC_tpm_protein_coding_clean.csv")
    df_rnaseq.set_index('gene_name',inplace=True)
    df_rnaseq = np.log2(df_rnaseq + 1)
    
print(df_rnaseq.shape)
df_rnaseq.head()
```

### sample_id：样本编号

```python
df_rnaseq_interest = df_rnaseq.loc[['KLRB1'], :].T # ,'ALDH1','OCT4','c-Myc',
df_rnaseq_interest['sample_id'] = [x[:15] for x in df_rnaseq_interest.index]
# df_rnaseq_interest.reset_index(drop=True, inplace=True)
print(f"样本数={len(df_rnaseq_interest['sample_id'].unique())}")
df_rnaseq_interest.head()
```

### bcr_patient_barcode：患者ID

```python
df_rnaseq_interest['bcr_patient_barcode'] = df_rnaseq_interest.index.map(lambda x: x[:12])
df_rnaseq_interest.head()
```

```python
print(f"病人数={len(df_rnaseq_interest['bcr_patient_barcode'].unique())}, 样本数={len(df_rnaseq_interest['sample_id'].unique())}")
```

### 临床数据中病人数比DKK1病人数多

```python
set(df_clinical_cBioPortal['patientId'].unique()) - set(df_rnaseq_interest['bcr_patient_barcode'].unique())
```

### Sample：编号01-09为癌症组织，10-19表示正常对照

```python
df_rnaseq_interest['Sample'] = df_rnaseq_interest.index.map(lambda x: x[13:15]).astype(int)
df_rnaseq_interest['Sample'].unique()
```

### Vial：A表示冰冻组织; B表示福尔马林固定石蜡包埋组织。
已被证明用于测序分析的效果不佳，所以不建议使用类似-01B的样本数据

```python
df_rnaseq_interest['Vial'] = df_rnaseq_interest.index.map(lambda x: x[15])
df_rnaseq_interest['Vial'].unique()
```

### Portion：同属于一个患者组织的不同部分的顺序编号，同一组织会分割为100-120mg的部分，分别使用

```python
df_rnaseq_interest['Portion'] = df_rnaseq_interest.index.map(lambda x: x[17:19]).astype(int)
df_rnaseq_interest['Portion'].unique()
```

### Analyte：分析的分子类型。
对RNA数据来说，Analyte为R的优先级最该，其次是R和T，而对于DNA层面的分析来说，D的优先级最高。

```python
df_rnaseq_interest['Analyte'] = df_rnaseq_interest.index.map(lambda x: x[19])
df_rnaseq_interest['Analyte'].unique()
```

```python
df_rnaseq_interest.head()
```

## 对于一个患者有多个样本的进行筛选，最后只选一个样本进行分析

### Sample：编号01-09为癌症组织（建议选择）

```python
df_rnaseq_interest['Sample'].unique()
```

```python
df_rnaseq_interest = df_rnaseq_interest[df_rnaseq_interest['Sample'] < 10].copy()
df_rnaseq_interest['Sample'].unique()
```

```python
df_rnaseq_interest.shape
```

```python
print(f"病人数={len(df_rnaseq_interest['bcr_patient_barcode'].unique())}, 样本数={len(df_rnaseq_interest['sample_id'].unique())}")
```

```python
df_rnaseq_interest['bcr_patient_barcode'].value_counts().head()
```

### 如果bcr_patient_barcode，有多个DKK1表达，选择Sample最小的那个

```python
first_Sample_no = df_rnaseq_interest.groupby('bcr_patient_barcode')['Sample'].transform('min')
is_first_Sample = df_rnaseq_interest['Sample'] == first_Sample_no
```

```python
df_rnaseq_interest = df_rnaseq_interest.loc[is_first_Sample, ['sample_id', 'KLRB1']]
df_rnaseq_interest.shape
```

# DKK1和临床数据合并

```python
df_clinical_cBioPortal_KLRB1 = pd.merge(df_clinical_cBioPortal, df_rnaseq_interest, left_on='sampleId', right_on='sample_id', how='inner')
print(df_clinical_cBioPortal_KLRB1.shape)
df_clinical_cBioPortal_KLRB1.set_index(keys='patientId', inplace=True)
df_clinical_cBioPortal_KLRB1.head()
```

```python
os.makedirs('../../data/TCGA-LIHC/clean',exist_ok=True)
df_clinical_cBioPortal_KLRB1.to_csv('../../data/TCGA-LIHC/clean/TCGA-LIHC_clinical_cBioPortalData_KLRB1.csv')
```
